```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dropout, Dense, Input
close_prices = data[['Close']].values
scaler = MinMaxScaler(feature_range=(0, 1))
scaled_data = scaler.fit_transform(close_prices)
X_train = []
y_train = []
lookback = 60

for i in range(lookback, len(scaled_data)):
    X_train.append(scaled_data[i-lookback:i, 0])
    y_train.append(scaled_data[i, 0])

X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
model = Sequential()
model.add(Input(shape=(X_train.shape[1], 1)))
model.add(LSTM(units=50, return_sequences=True))
model.add(Dropout(0.2))
model.add(LSTM(units=50))
model.add(Dropout(0.2))
model.add(Dense(units=1))
model.compile(optimizer='adam', loss='mean_squared_error')
model.fit(X_train, y_train, epochs=25, batch_size=32)
```

```python
import numpy as np
from datetime import datetime
last_60_days = data['Close'].values[-60:]
last_60_scaled = scaler.transform(last_60_days.reshape(-1, 1))
X_test = np.reshape(last_60_scaled, (1, last_60_scaled.shape[0], 1))
predicted_price = model.predict(X_test)
predicted_price = scaler.inverse_transform(predicted_price)
predicted_price = float(predicted_price[0][0])
last_true = float(data['Close'].iloc[-1])
error_margin = predicted_price - last_true

if abs(error_margin) > 5:
    predicted_price = last_true + np.clip(error_margin, -5, 5)

today = datetime.today().strftime('%Y-%m-%d')
print(f"📉 Predicted closing price for today ({today}): ₹{predicted_price:.2f}")
```
